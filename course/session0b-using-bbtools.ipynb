To make it easier for us to import Bitbrain's data into MNE structures, we've developed the `bbtools` library, a set of handy functions we found us constantly using in our projects.

So instead of going through the first notebook to import some Bitbrain's `csv` EEG data and wasting dozens of lines of code, we can just write:

```
bbt.read_csv("EEG.csv", ["Fp1", "Fp2", "C2", "C4"])
```

and we'll get the same MNE `raw` object.

````python
import bbtools as bbt
````

````python
# Check the README in the `/data` folder to read docummentation on each dataset.
eeg_file = "data/02-08-2020-jv-baseline-activity-no-labels-BBT-E16-AAA005-2020-02-08_12-59-21/EEG.csv"
electrodes_used = ['Fp1', 'Fp2', 'F3', 'F4', 'C1', 'C3', 'C2', 'C4', 'CP1', 'CP3', 'CP2', 'CP4', 'Cz', 'O1', 'O2', 'Pz']

raw = bbt.read_csv(eeg_file, electrodes_used)
````

````python
raw.plot(scalings = 'auto')
````
